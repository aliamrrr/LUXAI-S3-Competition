# Bot1 vs Bot2 Evaluation

```python
from luxai_s3.wrappers import LuxAIS3GymEnv
import numpy as np

def evaluate_agents(agent_1_cls, agent_2_cls, seed=42, games_to_play=3):
    env = LuxAIS3GymEnv(numpy_output=True)
    obs, info = env.reset(seed=seed)

    player_0 = agent_1_cls("player_0", info["params"])
    player_1 = agent_2_cls("player_1", info["params"])

    total_score_0 = 0
    total_score_1 = 0
    wins_0 = 0
    wins_1 = 0
    game_lengths = []
    total_steps = 0

    total_energy_0 = 0
    total_energy_1 = 0
    energy_samples = 0

    for i in range(games_to_play):
        obs, info = env.reset()
        game_done = False
        step = 0

        while not game_done:
            actions = {
                "player_0": player_0.act(step=step, obs=obs["player_0"]),
                "player_1": player_1.act(step=step, obs=obs["player_1"])
            }

            obs, rewards, terminated, truncated, info = env.step(actions)
            dones = {k: terminated[k] | truncated[k] for k in terminated}

            total_score_0 += rewards["player_0"]
            total_score_1 += rewards["player_1"]

            total_energy_0 += sum(obs["player_0"]["units"]["energy"]) / (sum(obs["player_0"]["units_mask"]) + 1)
            total_energy_1 += sum(obs["player_1"]["units"]["energy"]) / (sum(obs["player_1"]["units_mask"]) + 1)
            energy_samples += 1

            if dones["player_0"] or dones["player_1"]:
                game_done = True
                game_lengths.append(step)
                total_steps += step

                # winner
                if rewards["player_0"] > rewards["player_1"]:
                    wins_0 += 1
                elif rewards["player_1"] > rewards["player_0"]:
                    wins_1 += 1

            step += 1

    env.close()

    # Final stats
    avg_score_0 = total_score_0 / games_to_play
    avg_score_1 = total_score_1 / games_to_play
    win_rate_0 = wins_0 / games_to_play
    win_rate_1 = wins_1 / games_to_play
    avg_game_length = sum(game_lengths) / games_to_play
    points_per_step_0 = total_score_0 / (total_steps + 1)
    points_per_step_1 = total_score_1 / (total_steps + 1)
    avg_energy_0 = total_energy_0 / energy_samples
    avg_energy_1 = total_energy_1 / energy_samples

    avg_energy_0 = np.mean(avg_energy_0)
    avg_energy_1 = np.mean(avg_energy_1)

    return {
        "avg_score_0": avg_score_0,
        "avg_score_1": avg_score_1,
        "win_rate_0": win_rate_0,
        "win_rate_1": win_rate_1,
        "avg_game_length": avg_game_length,
        "points_per_step_0": points_per_step_0,
        "points_per_step_1": points_per_step_1,
        "avg_energy_0": avg_energy_0,
        "avg_energy_1": avg_energy_1
    }
```

# Multiple agents evaluation leaderboard

```python
import itertools
import pandas as pd

def evaluate_multiple_agents(agent_classes, seed=42, games_per_matchup=3, ranking_metric="Total Wins"):
    results = {agent: {"total_games": 0, "total_wins": 0, "total_score": 0, "total_steps": 0, "total_energy": 0, "total_win_rate": 0}
               for agent in agent_classes.keys()}

    matchups = list(itertools.combinations(agent_classes.keys(), 2))

    for agent_1_name, agent_2_name in matchups:
        agent_1_cls = agent_classes[agent_1_name]
        agent_2_cls = agent_classes[agent_2_name]

        print(f"ðŸ”„ Match: {agent_1_name} vs {agent_2_name}")

        # ExÃ©cuter les matchs
        match_result = evaluate_agents(agent_1_cls, agent_2_cls, seed=seed, games_to_play=games_per_matchup)
        print(match_result)

        # DÃ©terminer la victoire de la confrontation
        if match_result["win_rate_0"] > match_result["win_rate_1"]:
            results[agent_1_name]["total_wins"] += 1
        elif match_result["win_rate_1"] > match_result["win_rate_0"]:
            results[agent_2_name]["total_wins"] += 1

        # Ajouter le win rate global (accumulation)
        results[agent_1_name]["total_win_rate"] += match_result["win_rate_0"]
        results[agent_2_name]["total_win_rate"] += match_result["win_rate_1"]

        # Ajouter les autres statistiques
        results[agent_1_name]["total_games"] += 1  # Une confrontation compte comme 1
        results[agent_2_name]["total_games"] += 1

        results[agent_1_name]["total_score"] += match_result["avg_score_0"]
        results[agent_2_name]["total_score"] += match_result["avg_score_1"]

        results[agent_1_name]["total_steps"] += match_result["avg_game_length"]
        results[agent_2_name]["total_steps"] += match_result["avg_game_length"]

        results[agent_1_name]["total_energy"] += match_result["avg_energy_0"]
        results[agent_2_name]["total_energy"] += match_result["avg_energy_1"]

    # Construire le leaderboard
    leaderboard = []
    for agent, stats in results.items():
        leaderboard.append({
            "Agent": agent,
            "Total Wins": stats["total_wins"],  # Nombre total de victoires par confrontation
            "Win Rate": stats["total_win_rate"] / stats["total_games"],  # Moyenne des win rates
            "Avg Score": stats["total_score"] / stats["total_games"],
            "Avg Steps": stats["total_steps"] / stats["total_games"],
            "Avg Energy": stats["total_energy"] / stats["total_games"]
        })

    df_leaderboard = pd.DataFrame(leaderboard)

    # Trier selon la mÃ©trique clÃ©
    df_leaderboard = df_leaderboard.sort_values(by=ranking_metric, ascending=False)

    return df_leaderboard
```
